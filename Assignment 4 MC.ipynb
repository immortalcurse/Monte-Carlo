```python
import random
import math
import matplotlib.pyplot as plt
```

```python
def monte_carlo_pi(trials):
    inside_circle = 0
    for _ in range(trials):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        distance = math.sqrt(x**2 + y**2)
        if distance <= 1:
            inside_circle += 1
    return (inside_circle / trials) * 4
```

```python
def calculate_statistics(iterations, trials):
    results = []
    for _ in range(iterations):
        pi_estimate = monte_carlo_pi(trials)
        results.append(pi_estimate)
    mean = sum(results) / iterations
    variance = sum((x - mean) ** 2 for x in results) / iterations
    std_dev = math.sqrt(variance)
    return mean, std_dev, results
```

```python
# Number of trials in each Monte Carlo simulation
N = 1000

# Number of iterations for accumulating statistics
iterations = 200

# Calculate mean and standard deviation for 20 iterations of N trials each
mean_pi, std_dev_pi,pi_estimates = calculate_statistics(iterations, N)

print("Mean Estimate of π:", mean_pi)
print("Standard Deviation of Estimate:", std_dev_pi)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(pi_estimates, marker='o', linestyle='-', color='b')
plt.axhline(y=math.pi, color='r', linestyle='--', label='True π')
plt.xlabel('Iteration')
plt.ylabel('Estimated π')
plt.title('Monte Carlo Estimation of π')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import random
import math
import matplotlib.pyplot as plt

# Part (a)
def monte_carlo_pi(trials):
    inside_circle = 0
    for _ in range(trials):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        distance = math.sqrt(x**2 + y**2)
        if distance <= 1:
            inside_circle += 1
    return (inside_circle / trials) * 4

# Number of trials in each Monte Carlo simulation for part (a)
N_a = 100000
iterations_a = 20

# Perform Monte Carlo simulation for part (a) and calculate mean and standard deviation
pi_estimates = [monte_carlo_pi(N_a) for _ in range(iterations_a)]
mean_pi = sum(pi_estimates) / iterations_a
std_dev_pi = math.sqrt(sum((x - mean_pi) ** 2 for x in pi_estimates) / iterations_a)

print("(a) Mean Estimate of π:", mean_pi)
print("(a) Standard Deviation of Estimate:", std_dev_pi)

N_values = list(range(5, 100001, 20))
pi_estimates_b = []

# Perform Monte Carlo simulation for part (b) for varying N values
for N_b in N_values:
    pi_estimate = monte_carlo_pi(N_b)
    pi_estimates_b.append(pi_estimate)

# Plot the variation of π as a function of N for part (b)
plt.figure(figsize=(10, 6))
plt.plot(N_values, pi_estimates_b, linestyle='-', color='b')
plt.axhline(y=math.pi, color='r', linestyle='--', label='True π')
plt.xlabel('Number of Trials (N)')
plt.ylabel('Estimated π')
plt.title('Monte Carlo Estimation of π vs. Number of Trials')
plt.legend()
plt.grid(True)
plt.show()
```
